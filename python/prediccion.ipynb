```python
import pandas as pd
df = pd.read_excel('df_historic_order_demand.xlsx')
df.dropna(inplace=True)
df['fecha'] = pd.to_datetime(df['mes_anio'], format='%m-%Y')
df = df.sort_values(by='fecha')
df.set_index('fecha', inplace=True)
df.drop(columns=['mes_anio'], inplace=True)
df.head()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

df_result = pd.DataFrame(columns=['cliente', 'mes_anio', 'order_demand'])
for i in set(df['cliente'].values.tolist()):
    model = SARIMAX(df[df['cliente'] == i]['order_demand'], seasonal_order=(1, 1, 1, 12))
    model_fit = model.fit()
    forecast_next_year = model_fit.forecast(steps=12)

    # Crear índice de fechas para las predicciones
    last_date = df[df['cliente'] == i].index[-1]
    forecast_index = pd.date_range(last_date + pd.DateOffset(months=1), periods=12, freq='MS')

    # Crear un DataFrame para las predicciones
    forecast_df = pd.DataFrame({
        'mes_anio': forecast_index,
        'order_demand': forecast_next_year,
        'cliente': i
    })

    # Concatenar las predicciones al DataFrame final
    df_result = pd.concat([df_result, forecast_df], ignore_index=True)

    # Combinar predicciones con las fechas
    forecast_df = pd.DataFrame(forecast_next_year)
    print(i, forecast_df)
```
